```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier

from sklearn.preprocessing import OrdinalEncoder
from sklearn.preprocessing import StandardScaler
```

Data preprocessing and data uploaded

```python
df = pd.read_csv('iris.csv')
df
```

Initialize the feature cols and target col

```python
X = df.drop(columns = 'Species')
Y = df['Species']
X
```

split the data into train and test data

```python
xtrain, xtest, ytrain, ytest = train_test_split(X, Y, test_size= 0.2, random_state= 2)
```

create the object of the model to train data and predict the model performance

```python
knn = KNeighborsClassifier(n_neighbors = 5)
knn.fit(xtrain, ytrain)
y_pred = knn.predict(xtest)
y_pred
```

random input data pass o check the model's performance

```python
input_data = [[146,  6.7,	3.0,  5.2,	2.3]]
y_pred_input = knn.predict(input_data)
y_pred_input
```

visualize the data

```python
xtrain = xtrain.drop(columns = 'Id')
sns.barplot(xtrain)
plt.title('Before Train Model')
```

```python
xtest = xtest.drop(columns = 'Id')
sns.barplot(xtest)
plt.title(' After Train Model')
```

```python

sns.scatterplot(xtrain)
plt.title(' Before Train Model')
```

```python

sns.scatterplot(xtest)
plt.title(' After Train Model')
```

```python

sns.pairplot(xtrain)
plt.title('Before Train Model')
```

```python

sns.pairplot(xtest)
plt.title('After Train Model')
```
