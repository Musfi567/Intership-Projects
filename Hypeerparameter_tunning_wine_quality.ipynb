Filled the missing values

```python
df['fixed acidity'] = df['fixed acidity'].mean()
df.fillna(df['fixed acidity'], inplace = True)


df['volatile acidity'] = df['volatile acidity'].mean()
df.fillna(df['volatile acidity'], inplace = True)


df['citric acid'] = df['citric acid'].mean()
df.fillna(df['citric acid'], inplace = True)


df['residual sugar'] = df['residual sugar'].mean()
df.fillna(df['residual sugar'], inplace = True)


df['chlorides'] = df['chlorides'].mean()
df.fillna(df['chlorides'], inplace = True)


df['pH'] = df['pH'].mean()
df.fillna(df['pH'], inplace = True)


df['sulphates'] = df['sulphates'].mean()
df.fillna(df['sulphates'], inplace = True)
```

transform the type col , categorical to numerical by onehot encoder

```python
df['type'].unique()
```

```python
ohe = OneHotEncoder(sparse_output = False)
df['type'] = ohe.fit_transform(df[['type']])
df
```

initialize the feature cols and target col

```python
X = df.drop(columns = 'quality')
Y = df['quality']
```

split the data into train and test data

```python
xtrain, xtest, ytrain, ytest = train_test_split(X, Y, test_size= 0.2, random_state= 2)
```

Create the model's object to train the model then predict the target

different   model's can be train

Model can train by LinearRegressor

```python
lr = LinearRegression()
lr.fit(xtrain, ytrain)
y_pred_lr = lr.predict(xtest)
y_pred_lr
```

Model can train by k nearest neighbor regressor

```python
knn = KNeighborsRegressor(n_neighbors= 5)
knn.fit(xtrain, ytrain)
y_pred_knn = knn.predict(xtest)
y_pred_knn
```

Model can train by Decision Tree Regressor

```python
dtr = DecisionTreeRegressor()
dtr.fit(xtrain, ytrain)
y_pred_dtr = dtr.predict(xtest)
y_pred_dtr
```

Model can train by Random Froest Regressor

```python
rfr = RandomForestRegressor()
rfr.fit(xtrain, ytrain)
y_pred_rfr = rfr.predict(xtest)
y_pred_rfr
```

Check the cross validation score for all the different model's

```python
def cross_val(model):
    cross_val_data = cross_val_score(model , X, Y, cv= 10)
    cross_val_data_mean = cross_val_data.sum()/len(cross_val_data)
    print(f"The cross validation score of the {model} is", cross_val_data_mean)
```

Hyper Parameter Tunning 

```python
hyperparameter = {

        'lr' : {
            'n_jobs' : [-1, None]
        },

        'knn' : {
            'n_neighbors' : [5,8, 10],
            'leaf_size' : [10, 15, 20, 25, 30]
        },

        'dtr' : {
            'max_depth' : [1,2,3,4],
            'max_leaf_nodes' : [4,5,6,7,8],
            'min_samples_split' : [2,3,4],
            'random_state' : [2, 3,4, None]
        },

        'rfr' : {
            'n_estimators' : [10,50,80,100],
            'max_depth' : [1,2,3,4],
            'max_leaf_nodes' : [4,5,6,7,8],
            'min_samples_split' : [2,3,4],
            'random_state' : [2, 3,4, None]
        }

}
```

```python
steps = {
    'lr' : LinearRegression(),
    'knn' : KNeighborsRegressor(),
    'dtr' : DecisionTreeRegressor(),
    'rfr' : RandomForestRegressor()
}
```

hyper parameter tunning by Grid search cv

```python
for value, key in steps.items():
    gscv = GridSearchCV(estimator= key, param_grid= hyperparameter[value], cv = 5)
    gscv.fit(xtrain, ytrain)
    y_pred_gscv = gscv.predict(xtest)


    print(f"The grid search cv best score of the {key} is", gscv.best_score_)
    print(f"The grid search cv best parameter of the {key} is", gscv.best_params_)
```

hyper parameter tunning by randomized search cv

```python
for value, key in steps.items():
    rscv = RandomizedSearchCV(estimator= key, param_distributions = hyperparameter[value], cv = 5)
    rscv.fit(xtrain, ytrain)
    y_pred_gscv = rscv.predict(xtest)


    print(f"The grid search cv best score of the {key} is", rscv.best_score_)
    print(f"The grid search cv best parameter of the {key} is", rscv.best_params_)
```
