```python
import pandas as pd
import numpy as np

from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier

from sklearn.metrics import classification_report
```

Data Preprocessing and Data uploaded

```python
df = pd.read_csv('water_potability.csv')
df
```

```python
df.drop_duplicates(inplace= True)
```

```python
df.isnull().sum()
```

```python
df.shape
```

Filled the missing values

```python
df['ph'] = df['ph'].mean()
df.fillna(df['ph'], inplace = True)


df['Sulfate'] = df['Sulfate'].mean()
df.fillna(df['Sulfate'], inplace = True)


df['Trihalomethanes'] = df['Trihalomethanes'].mean()
df.fillna(df['Trihalomethanes'], inplace = True)
```

```python
df.isnull().sum()
```

Initialize the feature cols and target col

```python
X = df.drop(columns = 'Potability')
Y = df['Potability']
```

Split the data into train and test data

```python
xtrain, xtest, ytrain, ytest = train_test_split(X, Y, test_size= 0.2, random_state= 2)
```

Create the object model's to train the data and predict the model can correctlt work or not

```python
knn = KNeighborsClassifier(n_neighbors = 5)
knn.fit(xtrain, ytrain)
y_pred = knn.predict(xtest)
y_pred
```

Pass the Random input data to predict the model

```python
input_values = knn.predict([[8.316766,	214.373394,	22018.417441,	8.059332,	356.886136,	363.266516,	18.436524,	100.341674,	4.628771]])
input_values
```

Check the models performance by classification report

```python
performance_report = classification_report(ytest, y_pred, output_dict= True)
performance_report = pd.DataFrame(performance_report )
performance_report
```
